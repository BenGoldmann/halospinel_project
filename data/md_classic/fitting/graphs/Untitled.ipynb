```python
import matplotlib.pyplot as plt
import numpy as np

distance = np.linspace(0, 5, 500)

sc_cl = [(26156.48*np.exp(-i/0.2499)) for i in distance]
li_cl = [(385.14*np.exp(-i/0.3400)) for i in distance]
li_li = [(360.53*np.exp(-i/0.1610)) for i in distance]
cl_cl = [(1227.20*np.exp(-i/0.3210)-14.53/i**6) for i in distance]

plt.plot(distance, sc_cl)
plt.plot(distance, li_cl)
plt.plot(distance, li_li)
plt.plot(distance, cl_cl)

plt.ylabel('Energy (eV)')
plt.xlabel('Distance ($\mathrm{\AA}$)')

plt.legend(['Sc-Cl', 'Li-Cl', 'Li-Li', 'Cl-Cl'])

plt.ylim(0, 60)

def make_square_axes(ax):
    """Make an axes square in screen units.

    Should be called after plotting.
    """
    ax.set_aspect(1 / ax.get_data_ratio())

make_square_axes(plt.gca())

plt.savefig('potentials.jpg', bbox_inches='tight', dpi=500)
```
