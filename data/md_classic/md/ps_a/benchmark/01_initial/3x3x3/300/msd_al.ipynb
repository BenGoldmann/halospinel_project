# 600 K test

```python
import matplotlib.pyplot as plt
import numpy as np

# Import MSD files as lists:
lammps_step, lammps = np.loadtxt('super_al_msd_li.txt', skiprows=3, unpack=True)
ovito_step, ovito = np.loadtxt('msd_data.txt', skiprows=2, unpack=True)

steps = lammps_step / 1000000

# Plot MSD:   
plt.plot(steps, ovito, color='green')
plt.plot(steps, lammps)


plt.xlabel('time, ns', size = 12)
plt.ylabel('MSD, A^2', size = 12)
plt.title('MSD @600K', size=16)
```
