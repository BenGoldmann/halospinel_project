```python
import matplotlib.pyplot as plt
import numpy as np

# Import MSD files as lists:
li_step, li = np.loadtxt('300/super_al_msd_li.txt', skiprows=2, unpack=True)
sc_step, sc = np.loadtxt('300/super_al_msd_sc.txt', skiprows=2, unpack=True)
cl_step, cl = np.loadtxt('300/super_al_msd_cl.txt', skiprows=2, unpack=True)

steps = li_step / 1000000

li = li - li[1]
sc = sc - sc[1]
cl = cl - cl[1]

# Plot MSD:   
plt.plot(steps[1:], li[1:])
plt.plot(steps[1:], sc[1:])
plt.plot(steps[1:], cl[1:])

plt.xlabel('time, ns', size = 12)
plt.ylabel('MSD, A^2', size = 12)
plt.title('MSD @300K', size=16)

# Find diffusion coefficient:
li_slope, intercept = np.polyfit(steps[500:], li[500:], 1)
li_slope_standard_300 = float(li_slope * (10 ** (-11))) * 0.5
sc_slope, intercept = np.polyfit(steps[500:], sc[500:], 1)
sc_slope_standard_300 = float(sc_slope * (10 ** (-11))) * 0.5
cl_slope, intercept = np.polyfit(steps[500:], cl[500:], 1)
cl_slope_standard_300 = float(cl_slope * (10 ** (-11))) * 0.5

# Standard is for correction for A**2 and ns to SI base units
#
# Haven ratio on 0.5 is applied

print(f'The diffusion coefficient for Li @300k is {li_slope_standard_300} m^2/s')
print(f'The diffusion coefficient for Sc @300k is {sc_slope_standard_300} m^2/s')
print(f'The diffusion coefficient for Cl @300k is {cl_slope_standard_300} m^2/s')

li_cond_300 = (((1.602 * 10**(-19))**2 * (1.559 * 10**(28)) * li_slope_standard_300)/((1.381 * 10**(-23)) * 300) * 10**(-2))
sc_cond_300 = (((1.602 * 10**(-19))**2 * (4.585 * 10**(27)) * sc_slope_standard_300)/((1.381 * 10**(-23)) * 300) * 10**(-2))
cl_cond_300 = (((1.602 * 10**(-19))**2 * (2.934 * 10**(28)) * cl_slope_standard_300)/((1.381 * 10**(-23)) * 300) * 10**(-2))

print(f'The conductivity for Li @300k is {li_cond_300} S/cm')
print(f'The conductivity for Sc @300k is {sc_cond_300} S/cm')
print(f'The conductivity for Cl @300k is {cl_cond_300} S/cm')

# Conversion of diffusion coefficient to conductivity as follows:
# conducitivity = (charge**2 * charge carrier concentration * diffusion coefficient) / boltzmann constant * temperature
#
# charge is electron charge (sodium singularly charged), so 1.602 * 10**(-19) C
#
# charge carrier concentration calculated as number of Li/Sc/Cl ions in the simulation cell/cell dimensions
#
# diffusion coefficient calcualted as above fromm MSD plot slopes
#
# boltzmann constant = 1.381 * 10**(-23) J/K
#
# temperature varies with experiment
#
# !!! 10**(-2) added to get final results in S/cm as conventionally displayed

def make_square_axes(ax):
    """Make an axes square in screen units.

    Should be called after plotting.
    """
    ax.set_aspect(1 / ax.get_data_ratio())
    
plt.legend(labels = ['Li', 'Sc', 'Cl'])

make_square_axes(plt.gca())

plt.savefig('msd_300.jpg', bbox_inches='tight', dpi=500)
```
