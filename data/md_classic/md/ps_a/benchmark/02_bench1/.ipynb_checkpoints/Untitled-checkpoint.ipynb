```python
import matplotlib.pyplot as plt
import numpy as np

atoms = [1258, 3456, 18522, 27648]
atoms_extra = [27648, 39366]
speed = [ 0.36, 0.453, 1.20, 1.55]

linear_model=np.polyfit(atoms, speed, 1)
linear_model_fn=np.poly1d(linear_model)

print(linear_model_fn(39366))

plt.plot(atoms, speed)
plt.plot(atoms_extra, linear_model_fn(atoms_extra))
```
