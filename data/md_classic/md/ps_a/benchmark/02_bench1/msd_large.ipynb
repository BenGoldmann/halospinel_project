# Monoclinic 3x3x3

```python
import matplotlib.pyplot as plt
import numpy as np

# Import MSD files as lists:
step, mono3 = np.loadtxt('3x3x3/monoclinic_3x3x3_300_li.txt', skiprows=2, unpack=True)
step, tri3 = np.loadtxt('3x3x3/triclinic_3x3x3_300_li.txt', skiprows=2, unpack=True)
step, mono7 = np.loadtxt('7x7x7/monoclinic_7x7x7_300_li.txt', skiprows=2, unpack=True)
step, tri7 = np.loadtxt('7x7x7/triclinic_7x7x7_300_li.txt', skiprows=2, unpack=True)

steps = step / 1000000

mono3 = mono3 - mono3[1]
tri3 = tri3 - tri3[1]
mono7 = mono7 - mono7[1]
tri7 = tri7 - tri7[1]

# Plot MSD:   
plt.plot(steps[1:], mono3[1:])
plt.plot(steps[1:], tri3[1:])
plt.plot(steps[1:], mono7[1:])
plt.plot(steps[1:], tri7[1:])

plt.xlabel('time, ns', size = 12)
plt.ylabel('MSD, A^2', size = 12)
plt.title('MSD @300K', size=16)
plt.legend(['mono3', 'tri3', 'mono7', 'tri7'])

# Find diffusion coefficient:
mono3_slope, intercept = np.polyfit(steps, mono3, 1)
mono3_slope_standard_600 = float(mono3_slope * (10 ** (-11)))
tri3_slope, intercept = np.polyfit(steps, tri3, 1)
tri3_slope_standard_600 = float(tri3_slope * (10 ** (-11)))
mono7_slope, intercept = np.polyfit(steps[6000:], mono7[6000:], 1)
mono7_slope_standard_600 = float(mono7_slope * (10 ** (-11)))
tri7_slope, intercept = np.polyfit(steps[6000:], tri7[6000:], 1)
tri7_slope_standard_600 = float(tri7_slope * (10 ** (-11)))

print(f'The diffusion coefficient for monoclinic 3x3x3 @300k is {mono3_slope_standard_600} m^2/s')
print(f'The diffusion coefficient for triclinic 3x3x3 @300k is {tri3_slope_standard_600} m^2/s')
print(f'The diffusion coefficient for monoclinic 7x7x7 @300k is {mono7_slope_standard_600} m^2/s')
print(f'The diffusion coefficient for triclinic 7x7x7 @300k is {tri7_slope_standard_600} m^2/s')

mono3_cond = (((1.602 * 10**(-19))**2 * (1.559 * 10**(28)) * mono3_slope_standard_600)/((1.381 * 10**(-23)) * 600) * 10**(-2))
tri3_cond = (((1.602 * 10**(-19))**2 * (4.585 * 10**(27)) * tri3_slope_standard_600)/((1.381 * 10**(-23)) * 600) * 10**(-2))
mono7_cond = (((1.602 * 10**(-19))**2 * (2.934 * 10**(28)) * mono7_slope_standard_600)/((1.381 * 10**(-23)) * 600) * 10**(-2))
tri7_cond = (((1.602 * 10**(-19))**2 * (2.934 * 10**(28)) * tri7_slope_standard_600)/((1.381 * 10**(-23)) * 600) * 10**(-2))

print(f'The conductivity for monoclinic 3x3x3 @300k is {mono3_cond} S/cm')
print(f'The conductivity for triclinic 3x3x3 @300k is {tri3_cond} S/cm')
print(f'The conductivity for monoclinic 7x7x7 @300k is {mono7_cond} S/cm')
print(f'The conductivity for triclinic 7x7x7 @300k is {tri7_cond} S/cm')

# Conversion of diffusion coefficient to conductivity as follows:
# conducitivity = (charge**2 * charge carrier concentration * diffusion coefficient) / boltzmann constant * temperature
#
# charge is electron charge (sodium singularly charged), so 1.602 * 10**(-19) C
#
# charge carrier concentration calculated as number of Li/Sc/Cl ions in the simulation cell/cell dimensions
#
# diffusion coefficient calcualted as above fromm MSD plot slopes
#
# boltzmann constant = 1.381 * 10**(-23) J/K
#
# temperature varies with experiment
#
# !!! 10**(-2) added to get final results in S/cm as conventionally displayed
```
