# Organise data

```python
import re
import numpy as np



super_3_file = open('3_01_al/log.lammps')
super_4_file = open('4_01_al/log.lammps')
initial_file = open('3_05_al/log.lammps')
cu_file = open('3_05_cu/log.lammps')
drag_file = open('3_01_al_drag/log.lammps')
tchain_file = open('3_01_al_tchain/log.lammps')



super_3_line = super_3_file.readlines()
super_4_line = super_4_file.readlines()
initial_line = initial_file.readlines()
cu_line = cu_file.readlines()
drag_line = drag_file.readlines()
tchain_line = tchain_file.readlines()

super_3_data = super_3_line[173:274]
super_4_data = super_4_line[173:274]
initial_data = initial_line[170:221]
cu_data = cu_line[170:221]
drag_data = drag_line[173:274]
tchain_data = tchain_line[173:274]

time = []
time2 = []
time_scaled = []
time2_scaled = []

super_3_sorted = []
super_4_sorted = []
initial_sorted = []
cu_sorted = []
drag_sorted = []
tchain_sorted = []

super_3_energy = []
super_3_energy_unit = []
super_4_energy = []
super_4_energy_unit = []
initial_energy = []
initial_energy_unit = []
cu_energy = []
cu_energy_unit = []
drag_energy = []
drag_energy_unit = []
tchain_energy = []
tchain_energy_unit = []



for dat in super_3_data:
    split = re.split(' +', dat)
    super_3_sorted.append(float(split[2]))
    time.append(float(split[1]))
    
for dat in super_4_data:
    split = re.split(' +', dat)
    super_4_sorted.append(float(split[2]))
    
for dat in drag_data:
    split = re.split(' +', dat)
    drag_sorted.append(float(split[2]))
    
for dat in tchain_data:
    split = re.split(' +', dat)
    tchain_sorted.append(float(split[2]))
    
for dat in time:
    time_scaled.append(dat/1000)
    
for dat in initial_data:
    split = re.split(' +', dat)
    initial_sorted.append(float(split[2]))
    time2.append(float(split[1]))
    
for dat in cu_data:
    split = re.split(' +', dat)
    cu_sorted.append(float(split[2]))
    
for dat in time2:
    time2_scaled.append(dat/1000)

####################    
    
f_var_3 = 0
f_var_4 = 0
f_var_initial = 0
f_var_cu = 0
f_var_drag = 0
f_var_tchain = 0

for dat in super_3_sorted:
    diff = (dat - 300)**2
    f_var_3 = f_var_3 + diff
    
f_var_3 = np.sqrt(f_var_3/100)

print(f'3x3x3 | damp=0.1 | al - {f_var_3}')    


for dat in super_4_sorted:
    diff = (dat - 300)**2
    f_var_4 = f_var_4 + diff
    
f_var_4 = np.sqrt(f_var_4/100)
    
print(f'4x4x4 | damp=0.1 | al - {f_var_4}')  


for dat in drag_sorted:
    diff = (dat - 300)**2
    f_var_drag = f_var_drag + diff
    
f_var_drag = np.sqrt(f_var_drag/100)
    
print(f'3x3x3 | damp=0.1 | al | drag - {f_var_drag}')  


for dat in tchain_sorted:
    diff = (dat - 300)**2
    f_var_tchain = f_var_tchain + diff
    
f_var_tchain = np.sqrt(f_var_tchain/100)
    
print(f'3x3x3 | damp=0.1 | al | tchain - {f_var_tchain}')  
  
    
for dat in initial_sorted:
    diff = (dat - 300)**2
    f_var_initial = f_var_initial + diff
    
f_var_initial = np.sqrt(f_var_initial/50)
    
print(f'3x3x3 | damp=0.5 | al - {f_var_initial}')  
  
    
for dat in cu_sorted:
    diff = (dat - 300)**2
    f_var_cu = f_var_cu + diff
    
f_var_cu = np.sqrt(f_var_cu/50)
    
print(f'3x3x3 | damp=0.5 | cu - {f_var_cu}')  
    
###################    

for dat in super_3_data:
    split = re.split(' +', dat)
    super_3_energy.append(float(split[5]))
    
for dat in super_3_energy:
    super_3_energy_unit.append(dat/27)
    
for dat in super_4_data:
    split = re.split(' +', dat)
    super_4_energy.append(float(split[5]))
    
for dat in super_4_energy:
    super_4_energy_unit.append(dat/64)
    
for dat in drag_data:
    split = re.split(' +', dat)
    drag_energy.append(float(split[5]))
    
for dat in drag_energy:
    drag_energy_unit.append(dat/27)
    
for dat in tchain_data:
    split = re.split(' +', dat)
    tchain_energy.append(float(split[5]))
    
for dat in tchain_energy:
    tchain_energy_unit.append(dat/27)
    
for dat in initial_data:
    split = re.split(' +', dat)
    initial_energy.append(float(split[5]))
    
for dat in initial_energy:
    initial_energy_unit.append(dat/27)
    
for dat in cu_data:
    split = re.split(' +', dat)
    cu_energy.append(float(split[5]))
    
for dat in cu_energy:
    cu_energy_unit.append(dat/27)
```
