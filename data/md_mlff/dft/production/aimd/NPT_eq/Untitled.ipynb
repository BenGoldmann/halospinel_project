```python
file = open('vasp_out_1', 'r')
lines = file.readlines()

steps = []
temps = []

for l in lines:
    if 'T=' in l:
        split = l.split()
        steps.append(float(split[0]))
        temps.append(float(split[2]))
```

```python
import matplotlib.pyplot as plt

plt.plot(steps, temps)
```

```python
file = open('vasp_out_2', 'r')
lines = file.readlines()

for l in lines:
    if 'T=' in l:
        split = l.split()
        steps.append(float(split[0])+500)
        temps.append(float(split[2]))
```
