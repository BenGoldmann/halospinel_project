```python
from pymatgen.io.vasp.outputs import Xdatcar

data_1 = Xdatcar('XDATCAR_1')

structures_1 = data_1.structures
```

```python
a = []
b = []
c = []
alpha = []
beta = []
gamma = []

for s in structures_1:
    a.append(s.lattice.a)
    b.append(s.lattice.b)
    c.append(s.lattice.c)
    alpha.append(s.lattice.alpha)
    beta.append(s.lattice.beta)
    gamma.append(s.lattice.gamma)
```

```python
steps = []

for i in range(0, len(a)):
    steps.append(i)
```

```python
import matplotlib.pyplot as plt

plt.plot(steps, a)
plt.plot(steps, b)
plt.plot(steps, c)
```

```python
plt.plot(steps, alpha)
plt.plot(steps, beta)
plt.plot(steps, gamma)
```
