```python
import matplotlib.pyplot as plt

dataset = ['rdf_ml.txt', 'rdf_aimd.txt']

for ds in dataset:
    data = open(ds, 'r')
    lines = data.readlines()

    x = []
    y = []

    for line in lines[2:]:
        step = line.split()
        x.append(float(step[0]))
        y.append(float(step[1]))
    
    if max(y) > 6:
        y = [i/(max(y)/6) for i in y]
    
    plt.plot(x, y)

plt.xlabel('Pair separation distance', size=12)
plt.ylabel('g(r)', size=12)
plt.savefig('rdf_comp.jpg', bbox_inches='tight', dpi=500)
```
