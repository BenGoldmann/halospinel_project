```python
from pymatgen.io.vasp.outputs import Xdatcar
from pymatgen.io.vasp.outputs import Oszicar
import numpy as np
import matplotlib.pyplot as plt
```

```python
T = 600

xdatcar = Xdatcar(f'XDATCAR_{T}k')

structures = xdatcar.structures
```

```python
oszicar = Oszicar(f'OSZICAR_{T}k')

ionic_steps = oszicar.ionic_steps
```

```python
a_param = []
b_param = []
c_param = []
alpha_param = []
beta_param = []
gamma_param = []

for s in structures:
    a_param.append(s.lattice.a)
    b_param.append(s.lattice.b)
    c_param.append(s.lattice.c)
    alpha_param.append(s.lattice.alpha)
    beta_param.append(s.lattice.beta)
    gamma_param.append(s.lattice.gamma)

energies = []    
    
for i in ionic_steps:
    energies.append(i['E0'])
```

```python
step = np.arange(0, len(a_param))
time = step*2
```

```python
plt.plot(time, energies)
```

```python
print(f'a: {np.mean(a_param[500:])}')
print(f'b: {np.mean(b_param[500:])}')
print(f'c: {np.mean(c_param[500:])}')
print(f'alpha: {np.mean(alpha_param[500:])}')
print(f'beta: {np.mean(beta_param[1000:])}')
print(f'gamma: {np.mean(gamma_param[1000:])}')
```
