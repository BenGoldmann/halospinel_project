```python
a = [21.15767511190748, 21.370192236946686, 21.576909230616284, 21.875766682388058]
b = [21.13055727011369, 21.389061782127637, 21.818133094253675, 21.832302966136513]
c = [21.216893049321143, 21.318339073785115, 21.509022382625705, 22.08040086191277]
t = 600, 700, 800, 900
```

```python
import matplotlib.pyplot as plt

plt.plot(t, a)
plt.plot(t, b)
plt.plot(t, c)
```

```python
import numpy as np

a_slope, a_intercept = np.polyfit(t, a, 1)
b_slope, b_intercept = np.polyfit(t, b, 1)
c_slope, c_intercept = np.polyfit(t, c, 1)

plt.plot([600, 900], [600*a_slope+a_intercept, 900*a_slope+a_intercept])
plt.plot([600, 900], [600*b_slope+b_intercept, 900*b_slope+b_intercept])
plt.plot([600, 900], [600*c_slope+c_intercept, 900*c_slope+c_intercept])

ave_slope = np.mean([a_slope, b_slope, c_slope])
ave_intercept = np.mean([a_intercept, b_intercept, c_intercept])

#plt.plot([600, 900], [600*ave_slope+ave_intercept, 900*ave_slope+ave_intercept], color= 'black')
```

```python
T = 900

print(f'a: {T*a_slope+a_intercept}')
print(f'b: {T*b_slope+b_intercept}')
print(f'c: {T*c_slope+c_intercept}')
```
