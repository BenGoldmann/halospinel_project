```python
xdat1 = open('XDATCAR_1', 'r')
xdat2 = open('XDATCAR_2', 'r')
xdat3 = open('XDATCAR_mlff', 'r')

xdat1_lines = xdat1.readlines()
xdat2_lines = xdat2.readlines()
xdat3_lines = xdat3.readlines()

xdat_lines = xdat1_lines + xdat2_lines

a_params = []
a_params_mlff = []

for i in xdat_lines:
    if '20.' in i:
        split = i.split()
        if '20.' in split[0]:
            a_params.append(float(split[0]))
            
for i in xdat3_lines:
    if '20.' in i:
        split = i.split()
        if '20.' in split[0]:
            a_params_mlff.append(float(split[0]))
    elif '21.' in i:
        split = i.split()
        if '21.' in split[0]:
            a_params_mlff.append(float(split[0]))
            
percent = []
percent_mlff = []

for a in a_params:
    percent.append(((a-a_params[0])/a_params[0])*100)
    
for a in a_params_mlff:
    percent_mlff.append(((a-a_params_mlff[0])/a_params_mlff[0])*100)
```

```python
import matplotlib.pyplot as plt
import numpy as np

step = np.arange(0, len(a_params))
time = step*2

step2 = np.arange(0, len(a_params_mlff))
time2 = step2*2

plt.plot(time, percent)
plt.plot(time2, percent_mlff)
plt.xlabel('time (fs)')
plt.ylabel('Change in a parameter (%)')

plt.savefig('latparam.jpg', bbox_inches='tight', dpi=500)
```

```python
percent_mlff_moving_ave = []

index = np.arange(0, len(percent_mlff))

for i in index:
    iplus1 = i +1
    percent_mlff_moving_ave.append(np.mean(percent_mlff[:iplus1]))
    
plt.plot(time2, percent_mlff)
plt.plot(time2, percent_mlff_moving_ave)
```
