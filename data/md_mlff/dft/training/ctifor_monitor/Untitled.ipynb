```python
MLCX00 = ['MLCX00_1.txt', 'MLCX00_2.txt', 'MLCX00_3.txt', 'MLCX00_4.txt', 'MLCX00_5.txt']
MLCX01 = ['MLCX01_1.txt', 'MLCX01_2.txt', 'MLCX01_3.txt', 'MLCX01_4.txt', 'MLCX01_5.txt']
MLCX02 = ['MLCX02_1.txt', 'MLCX02_2.txt', 'MLCX02_3.txt', 'MLCX02_4.txt', 'MLCX02_5.txt']
MLCX03 = ['MLCX03_1.txt', 'MLCX03_2.txt', 'MLCX03_3.txt', 'MLCX03_4.txt', 'MLCX03_5.txt']

MLCX00_steps = []
MLCX00_values = []

MLCX01_steps = []
MLCX01_values = []

MLCX02_steps = []
MLCX02_values = []

MLCX03_steps = []
MLCX03_values = []

for f in MLCX00:
    data = open(f, 'r')
    lines = data.readlines()
    
    calc = int(f[7])
    
    for l in lines:
        split = l.split()
        MLCX00_values.append(float(split[2]))
        MLCX00_steps.append((int(split[1])+calc*5000)/500-10)
        
for f in MLCX01:
    data = open(f, 'r')
    lines = data.readlines()
    
    calc = int(f[7])
    
    for l in lines:
        split = l.split()
        MLCX01_values.append(float(split[2]))
        MLCX01_steps.append((int(split[1])+calc*5000)/500-10)
        
for f in MLCX02:
    data = open(f, 'r')
    lines = data.readlines()
    
    calc = int(f[7])
    
    for l in lines:
        split = l.split()
        MLCX02_values.append(float(split[2]))
        MLCX02_steps.append((int(split[1])+calc*5000)/500-10)
        
for f in MLCX03:
    data = open(f, 'r')
    lines = data.readlines()
    
    calc = int(f[7])
    
    for l in lines:
        split = l.split()
        MLCX03_values.append(float(split[2]))
        MLCX03_steps.append((int(split[1])+calc*5000)/500-10)
```

```python
import matplotlib.pyplot as plt

plt.plot(MLCX00_steps, MLCX00_values)
plt.plot(MLCX01_steps, MLCX01_values)
plt.plot(MLCX02_steps, MLCX02_values)
plt.plot(MLCX03_steps, MLCX03_values)

plt.xlabel('Training time (ps)')
plt.ylabel('ML_CTIFOR value')
plt.savefig('mlctifor.jpg', bbox_inches='tight', dpi=500)
```
