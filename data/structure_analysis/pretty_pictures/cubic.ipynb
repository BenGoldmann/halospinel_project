```python
import matplotlib.pyplot as plt
import matplotlib.patches as patches
from matplotlib.path import Path
from matplotlib.lines import Line2D
import matplotlib.patheffects as PathEffects
from site_plot import Site
import numpy as np
from matplotlib import cm
col = cm.get_cmap('Paired')

ticks = []

x_len = 22.08
y_len = 12.80

fig, ax = plt.subplots(figsize=(18,36))
ax.set_aspect('equal', adjustable='box')
ax.axis('off')

verts = [
   (0, 0),  # left, bottom
   (0, y_len),  # left, top
   (x_len, y_len),  # right, top
   (x_len, 0),  # right, bottom
   (0, 0),  # ignored
]

codes = [
    Path.MOVETO,
    Path.LINETO,
    Path.LINETO,
    Path.LINETO,
    Path.CLOSEPOLY,
]

path = Path(verts, codes)

ax.set_xticks(ticks)
ax.set_yticks(ticks)

ax.set_xlim(0-1, x_len+1)
ax.set_ylim(0-1, y_len+1)

patch = patches.PathPatch(path, facecolor='white', lw=2)
ax.add_patch(patch)

ccp_high = [[0,0.25], [0.5,0.75], [1,0.25], [0.16667,0.75], [0.33333,0.25], [0.66667,0.25], [0.83333,0.75]]
ccp_low = [[0,0.75], [0.5, 0.25], [1,0.75], [0.16667,0.25], [0.33333,0.75], [0.66667,0.75], [0.83333,0.25]]
oct_16d_sc_high = [[0.5,0.5]]
oct_16d_sc_low = [[0.16667,0]]
oct_16d_high = [[0,0], [0,1], [0.33333,0], [0.33333,1], [0.66667,0], [0.66667,1], [1,0], [1,1]]
oct_16d_low = [[0.5,1], [0.83333,0]]
oct_16c_high = [[0.16667,0.5], [0.83333,0.5]]
oct_16c_low = [[0,0.5], [0.33333,0.5], [0.66667,0.5], [1,0.5]]
oct_16c_low_mix = [[0.16667,1], [0.5,0], [0.83333,1]]
tet_8a_high = [[0.16667,0.625], [0.83333,0.625]]
tet_8a_low = [[0.5,0.375]]
tet_8b_high = [[0.5,0.125]]
tet_8b_low = [[0.5,0.875]]
tet_48f_vhigh = [[0,0.625], [1,0.625]]
tet_48f_vhigh_mix = [[0.16667,0.125], [0.83333,0.125]]
tet_48f_vhigh_yellow = [[0.33333,0.625], [0.66667,0.625]]
tet_48f_high = [[0,0.375], [1,0.375]]
tet_48f_high_mix = [[0.16667,0.875], [0.83333,0.875]]
tet_48f_high_yellow = [[0.33333,0.375], [0.66667,0.375]]
tet_48f_low = [[0.66667,0.125], [1,0.125]]
tet_48f_low_yellow = [[0,0.125], [0.33333,0.125], [0.5,0.625]]
tet_48f_vlow = [[0,0.875], [0.33333,0.875], [0.66667,0.875], [0.83333,0.375], [1,0.875]]
tet_48f_vlow_yellow = [[0.16667,0.375]]

sites_list = []

for i in ccp_high:
    sites_list.append(Site(i[0], i[1], '0', 'ccp', '32e'))
for i in ccp_low:
    sites_list.append(Site(i[0], i[1], '1/2', 'ccp', '32e'))
for i in oct_16d_sc_high:
    sites_list.append(Site(i[0], i[1], '1/4', 'octahedral', '16d_sc', mix=True))
for i in oct_16d_sc_low:
    sites_list.append(Site(i[0], i[1], '3/4', 'octahedral', '16d_sc', mix=True))
for i in oct_16d_high:
    sites_list.append(Site(i[0], i[1], '1/4', 'octahedral', '16d'))
for i in oct_16d_low:
    sites_list.append(Site(i[0], i[1], '3/4', 'octahedral', '16d', mix=True))
for i in oct_16c_high:
    sites_list.append(Site(i[0], i[1], '1/4', 'octahedral', '16c', mix=True))
for i in oct_16c_low:
    sites_list.append(Site(i[0], i[1], '3/4', 'octahedral', '16c'))
for i in oct_16c_low_mix:
    sites_list.append(Site(i[0], i[1], '3/4', 'octahedral', '16c', mix=True))
for i in tet_8a_high:
    sites_list.append(Site(i[0], i[1], '5/8', 'tetrahedral', '8a'))
for i in tet_8a_low:
    sites_list.append(Site(i[0], i[1], '7/8', 'tetrahedral', '8a'))
for i in tet_8b_high:
    sites_list.append(Site(i[0], i[1], '1/8', 'tetrahedral', '8b'))
for i in tet_8b_low:
    sites_list.append(Site(i[0], i[1], '3/8', 'tetrahedral', '8b'))
for i in tet_48f_vhigh:
    sites_list.append(Site(i[0], i[1], '1/8', 'tetrahedral', '48f'))
for i in tet_48f_vhigh_mix:
    sites_list.append(Site(i[0], i[1], '1/8', 'tetrahedral', '48f', mix=True))
for i in tet_48f_vhigh_yellow:
    sites_list.append(Site(i[0], i[1], '1/8', 'tetrahedral', '48f', yellow=True))
for i in tet_48f_high:
    sites_list.append(Site(i[0], i[1], '3/8', 'tetrahedral', '48f'))
for i in tet_48f_high_mix:
    sites_list.append(Site(i[0], i[1], '3/8', 'tetrahedral', '48f', mix=True))
for i in tet_48f_high_yellow:
    sites_list.append(Site(i[0], i[1], '3/8', 'tetrahedral', '48f', yellow=True))
for i in tet_48f_low:
    sites_list.append(Site(i[0], i[1], '5/8', 'tetrahedral', '48f'))
for i in tet_48f_low_yellow:
    sites_list.append(Site(i[0], i[1], '5/8', 'tetrahedral', '48f', yellow=True))
for i in tet_48f_vlow:
    sites_list.append(Site(i[0], i[1], '7/8', 'tetrahedral', '48f'))
for i in tet_48f_vlow_yellow:
    sites_list.append(Site(i[0], i[1], '7/8', 'tetrahedral', '48f', yellow=True))

for i in sites_list:
    if i.site_type == 'ccp':
        depth = ax.text((i.position[0]*x_len+0.5), (i.position[1]*y_len+0.5), i.position[2], size=12)
        depth.set_path_effects([PathEffects.withStroke(linewidth=4, foreground='white')])
        species_label = ax.text((i.position[0]*x_len), (i.position[1]*y_len), 'Cl', color = 'white', ha='center', va='center', size=24)
        species_label.set_path_effects([PathEffects.withStroke(linewidth=2, foreground='black')])
        if i.position[2] == '0':
            ax.add_patch(patches.Circle(((i.position[0]*x_len), (i.position[1]*y_len)), 0.6, facecolor='tab:grey', edgecolor='black', linewidth=2))
        elif i.position[2] == '1/2':
            ax.add_patch(patches.Circle(((i.position[0]*x_len), (i.position[1]*y_len)), 0.6, facecolor='tab:grey', edgecolor='black', linewidth=2))
    elif i.site_type == 'octahedral':
        depth = ax.text((i.position[0]*x_len+0.5), (i.position[1]*y_len+0.5), i.position[2], size=12)
        depth.set_path_effects([PathEffects.withStroke(linewidth=4, foreground='white')])
        if i.site_id == '16d_sc':
            species_label = ax.text((i.position[0]*x_len), (i.position[1]*y_len), 'Sc', color = 'white', ha='center', va='center', size=24)
            species_label.set_path_effects([PathEffects.withStroke(linewidth=2, foreground='black')])
            if i.position[2] == '1/4':
                octahedron = ax.add_patch(patches.RegularPolygon(((i.position[0]*x_len), (i.position[1]*y_len)), 4, 0.7, facecolor=col(1), edgecolor='black', linewidth=2))
                octahedron.set_path_effects([PathEffects.withStroke(linewidth=7, foreground=col(3))])
            elif i.position[2] == '3/4':
                octahedron = ax.add_patch(patches.RegularPolygon(((i.position[0]*x_len), (i.position[1]*y_len)), 4, 0.7, facecolor=col(1), edgecolor='black', linewidth=2, hatch='xxx'))
                octahedron.set_path_effects([PathEffects.withStroke(linewidth=7, foreground=col(3))])
        elif i.site_id == '16d':
            species_label = ax.text((i.position[0]*x_len), (i.position[1]*y_len), 'Li', color = 'white', ha='center', va='center', size=24)
            species_label.set_path_effects([PathEffects.withStroke(linewidth=2, foreground='black')])
            if i.position[2] == '1/4':
                ax.add_patch(patches.RegularPolygon(((i.position[0]*x_len), (i.position[1]*y_len)), 4, 0.7, facecolor=col(1), edgecolor='black', linewidth=2))
            elif i.position[2] == '3/4':
                octahedron = ax.add_patch(patches.RegularPolygon(((i.position[0]*x_len), (i.position[1]*y_len)), 4, 0.7, facecolor=col(1), edgecolor='black', linewidth=2, hatch='xxx'))
                octahedron.set_path_effects([PathEffects.withStroke(linewidth=7, foreground=col(3))])
        elif i.site_id == '16c':
            if i.position[2] == '1/4':
                octahedron = ax.add_patch(patches.RegularPolygon(((i.position[0]*x_len), (i.position[1]*y_len)), 4, 0.7, facecolor=col(3), edgecolor='black', linewidth=2))
                octahedron.set_path_effects([PathEffects.withStroke(linewidth=7, foreground=col(1))])
            elif i.position[2] == '3/4':
                if i.mix == True:
                    octahedron = ax.add_patch(patches.RegularPolygon(((i.position[0]*x_len), (i.position[1]*y_len)), 4, 0.7, facecolor=col(3), edgecolor='black', linewidth=2, hatch='xxx'))
                    octahedron.set_path_effects([PathEffects.withStroke(linewidth=7, foreground=col(1))])
                else:
                    ax.add_patch(patches.RegularPolygon(((i.position[0]*x_len), (i.position[1]*y_len)), 4, 0.7, facecolor=col(3), edgecolor='black', linewidth=2, hatch='xxx'))
    elif i.site_type == 'tetrahedral':
        depth = ax.text((i.position[0]*x_len+0.3), (i.position[1]*y_len+0.3), i.position[2], size=12)
        depth.set_path_effects([PathEffects.withStroke(linewidth=4, foreground='white')])
        if i.site_id == '8a':
            ax.add_patch(patches.RegularPolygon(((i.position[0]*x_len), (i.position[1]*y_len)), 3, 0.5, facecolor=col(2), edgecolor='black', linewidth=2, hatch='xxx'))
        elif i.site_id == '8b':
            tetrahedron = ax.add_patch(patches.RegularPolygon(((i.position[0]*x_len), (i.position[1]*y_len)), 3, 0.5, facecolor=col(0), edgecolor='black', linewidth=2))
            tetrahedron.set_path_effects([PathEffects.withStroke(linewidth=7, foreground=col(2))])
        elif i.site_id == '48f':
            if i.position[2] in ['1/8', '3/8']:
                if i.mix == True:
                    tetrahedron = ax.add_patch(patches.RegularPolygon(((i.position[0]*x_len), (i.position[1]*y_len)), 3, 0.5, facecolor=col(2), edgecolor='black', linewidth=2))
                    tetrahedron.set_path_effects([PathEffects.withStroke(linewidth=7, foreground=col(0))])
                elif i.yellow == True:
                    ax.add_patch(patches.RegularPolygon(((i.position[0]*x_len), (i.position[1]*y_len)), 3, 0.5, facecolor=col(10), edgecolor='black', linewidth=2))
                else:
                    ax.add_patch(patches.RegularPolygon(((i.position[0]*x_len), (i.position[1]*y_len)), 3, 0.5, facecolor=col(2), edgecolor='black', linewidth=2))
            elif i.position[2] in ['5/8', '7/8']:
                if i.yellow == True:
                    ax.add_patch(patches.RegularPolygon(((i.position[0]*x_len), (i.position[1]*y_len)), 3, 0.5, facecolor=col(10), edgecolor='black', linewidth=2, hatch='xxx'))
                else:
                    ax.add_patch(patches.RegularPolygon(((i.position[0]*x_len), (i.position[1]*y_len)), 3, 0.5, facecolor=col(2), edgecolor='black', linewidth=2, hatch='xxx'))

style = "<|-|>, head_width=5, head_length=10"
kw1 = dict(arrowstyle=style, shrinkA=35, shrinkB=35, color="k", linewidth=2, connectionstyle="arc3,rad=.5")
kw1_inv = dict(arrowstyle=style, shrinkA=35, shrinkB=35, color="k", linewidth=2, connectionstyle="arc3,rad=-.5")
kw2 = dict(arrowstyle=style, shrinkA=45, shrinkB=45, color='k', linewidth=2, connectionstyle="arc3,rad=.1")
kw2_inv = dict(arrowstyle=style, shrinkA=45, shrinkB=45, color='k', linewidth=2, connectionstyle="arc3,rad=-.1")
kw3_squareup = dict(arrowstyle=style, shrinkA=30, shrinkB=20, color='k', linewidth=2, connectionstyle="arc3,rad=-1")
kw3_squareup_inv = dict(arrowstyle=style, shrinkA=30, shrinkB=20, color='k', linewidth=2, connectionstyle="arc3,rad=1")
kw3_triangleup = dict(arrowstyle=style, shrinkA=30, shrinkB=30, color='k', linewidth=2, connectionstyle="arc3,rad=-1")
kw3_triangleup_inv = dict(arrowstyle=style, shrinkA=30, shrinkB=30, color='k', linewidth=2, connectionstyle="arc3,rad=1")

a1 = patches.FancyArrowPatch((0*x_len, 0.5*y_len), (0.16667*x_len, 0.625*y_len), **kw2_inv)
a2 = patches.FancyArrowPatch((0*x_len, 0.5*y_len), (0*x_len, 0.375*y_len), **kw3_squareup)
a3 = patches.FancyArrowPatch((0*x_len, 0.5*y_len), (0*x_len, 0.875*y_len), **kw1)
a4 = patches.FancyArrowPatch((0*x_len, 0.875*y_len), (0.16667*x_len, 1*y_len), **kw2)
a5 = patches.FancyArrowPatch((0*x_len, 0.375*y_len), (0.16667*x_len, 0.5*y_len), **kw2)
a6 = patches.FancyArrowPatch((0.166667*x_len, 1*y_len), (0.33333*x_len, 0.875*y_len), **kw2)
a7 = patches.FancyArrowPatch((0.166667*x_len, 1*y_len), (0.16667*x_len, 0.625*y_len), **kw1)
a8 = patches.FancyArrowPatch((0.166667*x_len, 1*y_len), (0.16667*x_len, 0.875*y_len),**kw3_squareup)
a9 = patches.FancyArrowPatch((0.166667*x_len, 0.875*y_len), (0.16667*x_len, 0.5*y_len), **kw1)
a10 = patches.FancyArrowPatch((0.166667*x_len, 0.625*y_len), (0.16667*x_len, 0.5*y_len), **kw3_triangleup)
a11 = patches.FancyArrowPatch((0.166667*x_len, 0.625*y_len), (0.33333*x_len, 0.5*y_len), **kw2_inv)
a12 = patches.FancyArrowPatch((0.166667*x_len, 0.5*y_len), (0*x_len, 0.625*y_len), **kw2)
a13 = patches.FancyArrowPatch((0.166667*x_len, 0.5*y_len), (0.16667*x_len, 0.125*y_len), **kw1)
a14 = patches.FancyArrowPatch((0.33333*x_len, 0.875*y_len), (0.33333*x_len, 0.5*y_len), **kw1)
a15 = patches.FancyArrowPatch((0.33333*x_len, 0.5*y_len), (0.5*x_len, 0.375*y_len), **kw2)
a16 = patches.FancyArrowPatch((0.5*x_len, 0.375*y_len), (0.5*x_len, 0*y_len), **kw1)
a17 = patches.FancyArrowPatch((0.5*x_len, 0.375*y_len), (0.66667*x_len, 0.5*y_len), **kw2)
a18 = patches.FancyArrowPatch((0.5*x_len, 0*y_len), (0.66667*x_len, 0.125*y_len), **kw2_inv)
a19 = patches.FancyArrowPatch((0.66667*x_len, 0.5*y_len), (0.66667*x_len, 0.875*y_len), **kw1)
a20 = patches.FancyArrowPatch((0.66667*x_len, 0.875*y_len), (0.83333*x_len, 1*y_len), **kw2)
a21 = patches.FancyArrowPatch((0.66667*x_len, 0.5*y_len), (0.83333*x_len, 0.625*y_len), **kw2_inv)
a22 = patches.FancyArrowPatch((0.66667*x_len, 0.5*y_len), (0.83333*x_len, 0.375*y_len), **kw2)
a23 = patches.FancyArrowPatch((0.66667*x_len, 0.5*y_len), (0.66667*x_len, 0.125*y_len), **kw1_inv)
a24 = patches.FancyArrowPatch((0.83333*x_len, 1*y_len), (0.83333*x_len, 0.875*y_len), **kw3_squareup_inv)
a25 = patches.FancyArrowPatch((0.83333*x_len, 1*y_len), (1*x_len, 0.875*y_len), **kw2)
a26 = patches.FancyArrowPatch((0.83333*x_len, 1*y_len), (0.83333*x_len, 0.625*y_len), **kw1_inv)
a27 = patches.FancyArrowPatch((0.83333*x_len, 0.875*y_len), (0.83333*x_len, 0.5*y_len), **kw1_inv)
a28 = patches.FancyArrowPatch((0.83333*x_len, 0.875*y_len), (0.83333*x_len, 0.5*y_len), **kw1_inv)
a29 = patches.FancyArrowPatch((0.83333*x_len, 0.625*y_len), (1*x_len, 0.5*y_len), **kw2_inv)
a30 = patches.FancyArrowPatch((0.83333*x_len, 0.625*y_len), (0.83333*x_len, 0.5*y_len), **kw3_triangleup_inv)
a31 = patches.FancyArrowPatch((0.83333*x_len, 0.5*y_len), (1*x_len, 0.625*y_len), **kw2_inv)
a32 = patches.FancyArrowPatch((0.83333*x_len, 0.5*y_len), (1*x_len, 0.375*y_len), **kw2)
a33 = patches.FancyArrowPatch((0.83333*x_len, 0.5*y_len), (0.83333*x_len, 0.125*y_len), **kw1_inv)
a34 = patches.FancyArrowPatch((1*x_len, 0.875*y_len), (1*x_len, 0.5*y_len), **kw1)
a35 = patches.FancyArrowPatch((1*x_len, 0.5*y_len), (0.83333*x_len, 0.375*y_len), **kw2_inv)
a36 = patches.FancyArrowPatch((1*x_len, 0.5*y_len), (1*x_len, 0.375*y_len), **kw3_squareup_inv)
a37 = patches.FancyArrowPatch((1*x_len, 0.5*y_len), (1*x_len, 0.125*y_len), **kw1)

arrows = [a1, a2, a3, a4, a5, a6, a7, a8, a9, a10, a11, a12, a13, a14, a15, a16, a17, a18, a19, a20, a21, a22, a23, a24, a25, a26, a27, a28, a29, a30, a31, a32, a33, a34, a35, a36, a37]


for a in arrows:
    ax.add_patch(a)

plt.savefig('cubic.jpg', bbox_inches='tight', dpi=500)
```
